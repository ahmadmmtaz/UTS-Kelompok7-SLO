```
#Membaca file csv dengan memberikan nama tiap atribut (kolom) berdasarkan sumber dataset
dataset=pd.read_csv("breast-cancer.csv",names=['class','age','menopause','tumor size','inv-node','node-caps','deg-malig','breast','breast-quad','irradiat'])
```

```
pd.set_option("display.max_rows", None, "display.max_columns", None) # membaca dataset secara keseluruhan
dataset # menampilkan dataset
```

```
dataset=dataset[['age','menopause','tumor size','inv-node','node-caps','deg-malig','breast','breast-quad','irradiat','class']] #Menyesuaikan urutan atribut
dataset  # Menampilkan dataset
```

```
import numpy as np #library untuk memproses larik
import matplotlib.pyplot as plt #Library untuk memvisualisasikan data
```

```
#Pada dataset terdapat missing value
dataset.replace("?", np.nan, inplace = True) 
```

```
#Melihat jumlah duplikasi data pada dataset
dataset.duplicated().sum()
```

```
# Hilangkan duplikasi data
dataset.drop_duplicates(inplace=True)
dataset.duplicated().sum()
```

```
#Mengecek missing value pada variabel
dataset.isnull().sum()
```

```
#Melihat summary dataset
dataset.info()
```

```
dataset.dropna(inplace=True)
```

```
dataset.index=range(len(dataset))
dataset.describe(include='all') #CLEAN
```

```
X = dataset.iloc[:, 0:9].values #seleksi dengan menggunakan bilangan bulat
X #menampilkan x
```

```
y = dataset.iloc[:, 9].values #seleksi dengan menggunakan bilangan bulat
y #menampilkan y
```

```
# Mengonversi data biner (node-caps, breast,irradiat,class) menjadi numerik
from sklearn.preprocessing import LabelEncoder, OneHotEncoder 
from sklearn.compose import ColumnTransformer
labelencoder = LabelEncoder()
X[:, 4] = labelencoder.fit_transform(X[:, 4])
X[:, 6] = labelencoder.fit_transform(X[:, 4])
X[:, 8] = labelencoder.fit_transform(X[:, 8])
y = labelencoder.fit_transform(y)
X
```

```
dataset.head() #Menampilkan 5 baris pertama pada dataset
```

```
#Menampilkan seluruh data
pd.set_option("display.max_rows", None, "display.max_columns", None)
dataset
```

```
#Mendeklarasikan variabel prediksi dan variabel target
X = dataset.drop(['class'], axis=1) #Mendeklarasikan Variabel prediksi yang dimana merupakan seluruh atribut kecuali atribut class

y = dataset['class'] #Mendeklarasikan atribut class sebagai variabel target
```

```
# split X and y dalam training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42) # digunakan testing size sebesar 20% dengan ramdom state 42
```

```
# Mengecek tipe data di X_train

X_train.dtypes
```

```
# import category encoders

import category_encoders as ce
```

```
# encode variabel dengan pengkodean ordinal

encoder = ce.OrdinalEncoder(cols=['age', 'menopause', 'tumor size', 'inv-node', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```
#Menampilkan data 5 baris pertama pada X_train
X_train.head()
```

```
#Menampilkan data 5 baris pertama pada X_test
X_test.head()
```

```
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```
# membuat instance model DecisionTreeClassifier dengan kriteria entropy

clf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)


# fit the model
clf_en.fit(X_train, y_train)
```

```
#Memprediksi hasil Test set dengan kriteria entropi
y_pred_en = clf_en.predict(X_test) 
```

```
#Periksa skor akurasi dengan kriteria entropi
from sklearn.metrics import accuracy_score

print('Skor Akurasi Model dengan Kriteria Entropi: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

```
y_pred_train_en = clf_en.predict(X_train)

y_pred_train_en
```

```
print('Skor Akurasi Dari Training-set: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

```
# Menampilkan skor training and test set

print('Skor Training set : {:.4f}'.format(clf_en.score(X_train, y_train)))

print('Skor Test set : {:.4f}'.format(clf_en.score(X_test, y_test)))
```

```
3Memvisualisasikan decision tree models
plt.figure(figsize=(12,8)) 

from sklearn import tree

tree.plot_tree(clf_en.fit(X_train, y_train))
```

```
# Menampilkan Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```

```
#classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_en))
```
